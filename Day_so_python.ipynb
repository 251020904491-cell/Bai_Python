```python
daysonguyen=input("Nhap day so nguyen:")
danhsachdayso=daysonguyen.split()
danhsachchuoiso = list(map(int, danhsachdayso))
sochan=0
sole=0
for so in danhsachchuoiso:
   if so %2==0:
      sochan=sochan+so
   else :
      sole=sole+so
print("Tong so chan:",sochan)
print("Tong so le:",sole)
```
