```python
a = int(input("Nhập số nguyên thứ nhất: "))
b = int(input("Nhập số nguyên thứ hai: "))
c = int(input("Nhập số nguyên thứ ba: "))
max_value = a
if b > max_value:
  max_value = b
if c > max_value:
  max_value = c
print("Số lớn nhất là:", max_value)
```

```python
year = int(input("Nhập năm: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
  print(year, "là năm nhuận")
else:
  print(year, "không phải là năm nhuận")
```

```python
n = int(input("Nhập số nguyên dương: "))
tong = 0
for i in range(n, n + 1):
  tong += i
  print("Tổng từ 1 đến", n, "là", tong)
```

```python
am = 0
duong = 0
zero = 0

print("Nhập 10 số nguyên:")

for i in range(10):
    x = int(input(f"Nhập số thứ {i+1}; "))

    if x < 0:
        am += 1
    elif x > 0:
        duong += 1
    else:
        zero += 1

print("Số âm:", am)
print("Số dương:", duong)
print("Số bằng 0:", zero)
```

```python
n = int(input("Nhập một số từ 1 đến 9: "))

if 1 <= n <= 9:
    print(f"Bảng cửu chương {n}:")
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")
else:
     print("Vui lòng nhập số trong khoảng 1 đến 9.")
```

```python
n = int(input("Nhập một số n: "))

if n < 2:
    print(n, "không phải là số nguyên tố ")
else:
    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(n, "là số nguyên tố")
    else:
        print(n, "không phải số nguyên tố")
```
