```python
n = input("Nhap 3 so thuc vao de lap phuong trinh: ")
n = n.split()
a = float(n[0])
b = float(n[1])
c = float(n[2])
if a == 0:
    print("Phuong trinh vo nghiem")
else:
    delta =(b**2 - 4*a*c)>0
    if delta>0:
        print("Phuong trinh co 2 nghiem phan biet")
    elif delta == 0:
        print("Phuong trinh co nghiem kep")
    else:
        print('Phuong trinh vo nghiem trong tap so thuc')
```

```python
n = int(input("Nhap so nguyen duong n: "))

while n <= 0:
  print("n phai la so nguyen duong nhap lai.")
  n = int(input("Nhap lai so nguyen duong n: "))

tong = 0
i = 2

while i < n:
  tong += i
  i += 2
print("Tong cac so chan nho hon", n, "la:", tong)
```
